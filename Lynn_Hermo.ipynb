```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time 
from collections import defaultdict
```

```python
#set desired maximum page


page_url = []
max_page = 2
for i in range(1,max_page + 1):
    page_num = str(i)
    url = "https://www.hermo.my/skin-care/10-face-mask/page-" + page_num+ "'"
    page_url.append(url)

browser = webdriver.Chrome()
browser.get(url)
code = browser.page_source
soup = BeautifulSoup(code,'html.parser')



link2 = []
for url in page_url:

    soup_list2 = soup.findAll(class_ = 'hershop-product-cards _type-3 d-block')
    
    #find all href link for each product and dump it into a list
    link_list = [a['href'] for a in soup_list2]
    

    #edit the link and create a new product link list
    for link in link_list:
        link2.append("https://www.hermo.my" +link)



```

```python
#Create Empty list

Review = []
Product = []
Brand = []
Price = []

#loop through each product link
for url in link2:
    browser.get(url)
    code = browser.page_source
    soup = BeautifulSoup(code,'html.parser')
    time.sleep(3)
    Product.append(browser.find_element_by_class_name("mall-product-title").text)
    Brand.append(browser.find_element_by_class_name("mall-product-brand").text)
    Price.append(browser.find_element_by_class_name("column").text)
    
#create rating and review list for each product    
    rate = []
    rev = []
    soup_list = soup.find(id= 'test')
    review = soup_list.find_all(class_= 'line-height')
    rating = soup_list.find_all(class_= 'herelem-ratings')
    
#dump into the list    
    for item, star in zip(review,rating) :
        a = len(star.find_all('span',class_= 'ico star ico-star-full')) + len(star.find_all('span',class_= 'ico star ico-star-empty ico-star-full'))
        b = item.get_text()
        rate.append(a)
        rev.append(b)
        
#create dictionary and join rating n review
    res = defaultdict(list)
    for i, j in zip(rate, rev):
        res[i].append(j)
        
#append as a new list    
    Review.append(res)   


    
```

```python
#put into panda

import pandas as pd

Lynn = pd.DataFrame({
    'Product':Product,
    'Brand':Brand,
    'Price': Price,
    'Review':Review,  
    
})
```
